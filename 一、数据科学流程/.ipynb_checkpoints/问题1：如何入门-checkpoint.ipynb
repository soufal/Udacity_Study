```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../Data/survey-results-public.csv')
df.head()
```

```python
study = df['CousinEducation'].value_counts().reset_index()
study
```

一条数据记录会包含回答者选择的所有选项，因此需要对数据进行清洗和处理。

```python
#清洗数据，先将index列名改成更为合适的名称
study.rename(columns={'index': 'method', 'CousinEducation': 'count'}, inplace = True)
study.head()
```

创建一个包含所有方法选项的列表：
可以逐行读取数据，然后对method中的数据通过分号来进行分割。

```python
possible_vals = []
def possible_list(row_list):
    for i in row_list:
        if i not in possible_vals:
            possible_vals.append(i)
        else:
            pass
for index, row in study.iterrows():
    row_list = row['method'].split(';')
    possible_list(row_list)
#     print(row_list)
print(possible_vals)
```

```python
#统计各个method的频数
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df: 数据
    col1: 需要查看的列名
    col2：需要统计的count值
    look_for: possible_vals
    '''
    
    from collections import defaultdict
    new_df = defaultdict(int)
    for val in look_for:
        for idx in range(df.shape[0]):
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df
```

```python
study_df = total_count(study, 'method', 'count', possible_vals)
study_df
```
